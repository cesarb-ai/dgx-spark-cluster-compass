# 01 — First Spark: power, router (Cat6), SSH

**Milestone:** one DGX Spark is on your LAN and you can open a shell over SSH.

This is **before** the expensive QSFP cable between Sparks ([`03`](03_qsfp_interconnect_link_and_mtu.ipynb)). You only need:

- Power
- **Cat6 (or better) to the router/switch** — management path
- The IP the DHCP server gave you (or a static you assigned)

See also: [NVIDIA connect-two-sparks](https://github.com/NVIDIA/dgx-spark-playbooks/tree/main/nvidia/connect-two-sparks) for first-boot expectations.

## Checklist (do in the real world)

- [ ] Unbox, rack or bench, connect **power**.
- [ ] Connect **Ethernet** from Spark **management** port to your router/switch (Cat6+).
- [ ] Wait for boot; note **hostname** or use router DHCP lease table to find **IP**.
- [ ] From your laptop/desktop: `ssh youruser@<LAN_IP>` succeeds.
- [ ] Optional: set a **static DHCP reservation** so that IP stops moving.

**Green signal:** you have a repeatable SSH command for Spark **#1** on the **home/office LAN**.

```python
# Optional: from your laptop, test SSH reachability (edit host)
import shutil
import subprocess

SPARK1_LAN_IP = "192.168.0.28"  # <-- your first Spark on the router LAN
USER = "youruser"

ssh = shutil.which("ssh")
if not ssh:
    print("ssh not found in PATH")
else:
    r = subprocess.run(
        [ssh, "-o", "BatchMode=yes", "-o", "ConnectTimeout=5", f"{USER}@{SPARK1_LAN_IP}", "hostname"],
        capture_output=True,
        text=True,
    )
    print("✅ SSH ok:" if r.returncode == 0 else "❌ SSH failed:", (r.stdout or r.stderr).strip())
```
