# 01 — First Spark: power, router (Cat6), SSH

**Milestone:** Spark #1 is reachable over SSH on your management LAN.

## Do this first

1. Power on the first Spark.
2. Connect Spark management Ethernet to your router/switch (Cat6+).
3. Find the Spark LAN IP from your router DHCP leases.
4. Verify SSH works to that IP.

## Important for this notebook

If Jupyter is running on the Spark head (common for this repo), the SSH test cell below is optional and may fail unless key-based auth is configured.

See also: [NVIDIA connect-two-sparks](https://github.com/NVIDIA/dgx-spark-playbooks/tree/main/nvidia/connect-two-sparks).

## Quick checklist

- [ ] Spark powered on and connected to router/switch via management Ethernet.
- [ ] Spark LAN IP identified in DHCP leases.
- [ ] SSH command works from your admin machine: `ssh <user>@<LAN_IP>`.
- [ ] Optional: reserve a static DHCP lease for this Spark.

**Done when:** you can run the same SSH command reliably for Spark #1.

```python
# Optional check: SSH to Spark #1 (edit values)
# If this notebook runs on the head Spark, you can skip this cell.
# It is most useful when the kernel runs on your laptop/admin machine.
import shutil
import subprocess

SPARK1_LAN_IP = "192.168.0.28"  # Replace with Spark #1 LAN IP
USER = "cesarb-ai"               # Replace with your SSH username

ssh = shutil.which("ssh")
if not ssh:
    print("ssh not found in PATH")
else:
    r = subprocess.run(
        [ssh, "-o", "BatchMode=yes", "-o", "ConnectTimeout=5", f"{USER}@{SPARK1_LAN_IP}", "hostname"],
        capture_output=True,
        text=True,
    )
    if r.returncode == 0:
        print("✅ SSH ok:", r.stdout.strip())
    else:
        print("❌ SSH check failed.")
        print("If you can SSH manually with a password, ignore this result or remove BatchMode=yes.")
        print((r.stdout or r.stderr).strip())
```
