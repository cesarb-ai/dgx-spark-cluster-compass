```python
# ON THE HEAD SPARK — edit WORKER_IP
import subprocess

WORKER_IP = "10.0.0.2"

def run(c: list[str]) -> None:
    r = subprocess.run(c, capture_output=True, text=True)
    print(" ".join(c))
    print(r.stdout + r.stderr or "(no output)")

run(["ping", "-c", "3", "-W", "2", WORKER_IP])
print()
run(["ip", "route", "get", WORKER_IP])
```
